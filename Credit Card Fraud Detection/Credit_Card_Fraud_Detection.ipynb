## Importing Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("creditcard.csv")
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Class'].value_counts()
```

```python
normal = df[df.Class == 0]
```

```python
fraud = df[df.Class == 1]
```

```python
print(normal.shape)
```

```python
print(fraud.shape)
```

```python
normal.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
df.groupby("Class").mean()
```

```python
normal_sample = normal.sample(n = 492)
```

```python
mod_df = pd.concat([normal_sample, fraud], axis = 0)
```

```python
mod_df.head(10)
```

```python
mod_df['Class'].value_counts()
```

```python
mod_df.groupby("Class").mean()
```

```python
x = mod_df.drop(columns = "Class", axis = 1)
```

```python
y = mod_df["Class"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 2)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
x_train_prediction = model.predict(x_train)
```

```python
training_data_accuracy = accuracy_score(x_train_prediction, y_train) * 100
```

```python
print(f"Training Data Accuracy: {training_data_accuracy}%")
```

```python
x_test_prediction = model.predict(x_test)
```

```python
test_data_accuracy = accuracy_score(x_test_prediction, y_test) * 100
```

```python
print(f"Test Data Accuracy: {test_data_accuracy}%")
```
